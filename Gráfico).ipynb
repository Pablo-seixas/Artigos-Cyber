```python
# Importando as bibliotecas necessárias
import matplotlib.pyplot as plt
import numpy as np

# Dados
paises = ['Brasil', 'México', 'Venezuela', 'Colômbia', 'Chile']
tentativas_ataques = [23, 14, 10, 5, 4]  # em bilhões

# Criando o gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(paises, tentativas_ataques, color='blue')

# Adicionando títulos e rótulos
plt.title('Tentativas de Ataques Cibernéticos em 2023 (em bilhões)')
plt.xlabel('Países')
plt.ylabel('Tentativas de Ataques (em bilhões)')

# Mostrando o gráfico
plt.show()
```
